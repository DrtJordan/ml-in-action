```python
import numpy as np
import sys
from matplotlib.pyplot import plot
from matplotlib.pyplot import show
```

```python
N = 5
weights = np.ones(N)/N
print(weights)
```

```python
c = np.loadtxt('data.csv', delimiter=',',usecols=(6,), unpack=True)
print(c)
```

```python
sma = np.convolve(weights, c)[N-1:-N+1]
print(sma)
```

```python
t = np.arange(N-1,len(c))
plot(t, c[N-1:],lw=1.0)
plot(t, sma, lw=4.0)
show()
```
